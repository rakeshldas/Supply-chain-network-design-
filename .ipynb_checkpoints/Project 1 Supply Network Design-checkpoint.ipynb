# Business Problem

A company has factories, depots, and customers. Products move from factories to depots and then from depots to customers. The company wants to satisfy all customer demand at minimum total cost.

# This project covers:

Supply chain management
Network flow optimization
Depot opening decision
Transportation cost minimization
Capacity constraints
Model Logic in Simple Words

# The model decides:

How much quantity should move from each factory to each depot?

How much quantity should move from each depot to each customer?

Which depots should be opened?

# Objective:

Minimize total shipping cost + depot fixed opening cost

# Constraints:

Factory shipment cannot exceed factory capacity.

Depot flow going out cannot exceed depot capacity if depot is opened.

For every depot, quantity coming in must equal quantity going out.

Every customer demand must be fully satisfied.

# Project 1: Supply Network Design using Gurobi


## Business Objective

The objective of this project is to design a minimum-cost supply chain network.

The company has:

* Factories
* Depots
* Customers

The product flows in two stages:

```Factories → Depots → Customers```

The model decides:

1. how much quantity should be shipped from each factory to each depot,
2. how much quantity should be shipped from each depot to each customer,
3. which depots should be opened.

The objective is to minimize total supply chain cost.

---

# Step 1: Define Network Structure

## Business Extension

In this model, the supply chain has three layers.

| Layer     | Meaning                                   |
| --------- | ----------------------------------------- |
| Factories | Production or supply sources              |
| Depots    | Intermediate storage/distribution centers |
| Customers | Demand points                             |

The flow structure is:

```Factory → Depot → Customer```

## Sets

| Set | Meaning          | Elements                                       |
| --- | ---------------- | ---------------------------------------------- |
| F   | Set of factories | Factory 1, Factory 2                           |
| D   | Set of depots    | Depot 1, Depot 2, Depot 3                      |
| C   | Set of customers | Customer 1, Customer 2, Customer 3, Customer 4 |

```950 units```

# Step 5: Depot Fixed Opening Cost

## Business Extension

Opening a depot creates a fixed cost.

This cost is paid only if the depot is opened.

## Data

| ------- | -----------------: |
| Depot 1 |               1200 |
| Depot 2 |               1000 |
| Depot 3 |                900 |

The model should not open all depots automatically.

It should open a depot only if the transportation cost saving justifies the fixed opening cost.

# Step 6: Factory-to-Depot Transportation Cost

## Business Extension

Shipping from factories to depots has a per-unit cost.

The model decides how much quantity should be shipped on each route.

## Data

| From Factory | To Depot | Cost per Unit |
| ------------ | -------- | ------------: |
| Factory 1    | Depot 1  |             4 |
| Factory 1    | Depot 2  |             6 |
| Factory 1    | Depot 3  |             9 |
| Factory 2    | Depot 1  |             5 |
| Factory 2    | Depot 2  |             4 |
| Factory 2    | Depot 3  |             3 |

Let:

```x_fd = quantity shipped from factory f to depot d```

Then factory-to-depot shipping cost is:

```Factory-to-Depot Cost = Sum over all factories f and all depots d of: CostFD_fd × x_fd```

In simple words:

```Factory-to-Depot Cost = Cost per unit × Quantity shipped```

---

# Step 7: Depot-to-Customer Transportation Cost

## Business Extension

After products reach depots, they are shipped from depots to customers.

Each depot-to-customer route has a per-unit transportation cost.

The model decides how much quantity should be shipped from each depot to each customer.

## Data

| From Depot | To Customer | Cost per Unit |
| ---------- | ----------- | ------------: |
| Depot 1    | Customer 1  |             3 |
| Depot 1    | Customer 2  |             5 |
| Depot 1    | Customer 3  |             7 |
| Depot 1    | Customer 4  |             6 |
| Depot 2    | Customer 1  |             4 |
| Depot 2    | Customer 2  |             3 |
| Depot 2    | Customer 3  |             5 |
| Depot 2    | Customer 4  |             4 |
| Depot 3    | Customer 1  |             7 |
| Depot 3    | Customer 2  |             6 |
| Depot 3    | Customer 3  |             3 |
| Depot 3    | Customer 4  |             4 |

Let:

```y_dc = quantity shipped from depot d to customer c```

Then depot-to-customer transportation cost is:

```Depot-to-Customer Cost = Sum over all depots d and all customers c of: = CostDC_dc × y_dc```

In simple words:

```Depot-to-Customer Cost = Cost per unit × Quantity shipped```

---

# Step 8: Decision Variables

## Business Extension

The model needs to decide three types of values:

1. quantity shipped from factories to depots,
2. quantity shipped from depots to customers,
3. whether each depot should be opened or not.

These are called decision variables because Gurobi will choose their best values.

---

## Decision Variable 1: Factory-to-Depot Flow

Let:

```x_fd = quantity shipped from factory f to depot d```

where:

```f belongs to the set of factories F, d belongs to the set of depots D```

This variable is continuous and non-negative.

```x_fd must be greater than or equal to 0```

## Table Form

| Variable | Meaning                                    | Type       |
| -------- | ------------------------------------------ | ---------- |
| x_fd     | Quantity shipped from factory f to depot d | Continuous |

Example:

```x_Factory1_Depot2```

means:

```Quantity shipped from Factory 1 to Depot 2```

---

## Decision Variable 2: Depot-to-Customer Flow

Let:

```y_dc = quantity shipped from depot d to customer c```

where:

```d belongs to the set of depots D, c belongs to the set of customers C```

This variable is also continuous and non-negative.

```y_dc must be greater than or equal to 0```

## Table Form

| Variable | Meaning                                     | Type       |
| -------- | ------------------------------------------- | ---------- |
| y_dc     | Quantity shipped from depot d to customer c | Continuous |

Example:

```y_Depot2_Customer4```

means:

```Quantity shipped from Depot 2 to Customer 4```

---

## Decision Variable 3: Depot Opening Decision

Let:

```z_d = 1 if depot d is opened else z_d = 0 if depot d is closed```

where:

```d belongs to the set of depots D```

This is a binary decision variable.

```z_d can only be 0 or 1```

## Table Form

| Variable | Meaning                             | Type   |
| -------- | ----------------------------------- | ------ |
| z_d      | 1 if depot d is opened, 0 otherwise | Binary |

Example:

```z_Depot1 = 1```

means:

```Depot 1 is opened```

```z_Depot1 = 0```

means:

```Depot 1 is closed```

---

# Step 9: Objective Function

## Business Extension

The company wants to minimize total supply chain cost.

Total supply chain cost has three parts:

1. factory-to-depot transportation cost,
2. depot-to-customer transportation cost,
3. fixed depot opening cost.

So:

```Total Cost = Factory-to-Depot Cost + Depot-to-Customer Cost + Depot Opening Cost```

---

## Objective Function

```Minimize Z```

where:

```Z = Sum over all factories f and depots d of CostFD_fd × x_fd + Sum over all depots d and customers c of CostDC_dc × y_dc + Sum over all depots d of FixedCost_d × z_d```

---

## Explanation of Each Term

| Term                     | Meaning                                         |
| ------------------------ | ----------------------------------------------- |
| Sum of CostFD_fd × x_fd  | Total cost of shipping from factories to depots |
| Sum of CostDC_dc × y_dc  | Total cost of shipping from depots to customers |
| Sum of FixedCost_d × z_d | Total fixed cost of opened depots               |

---

## Objective in Simple Business Language

The model tries to find the cheapest supply chain plan by deciding:

* which depots to open,
* how much to ship from factories to depots,
* how much to ship from depots to customers.

The goal is:

```Minimize transportation cost and depot opening cost```

---

# Step 10: Factory Capacity Constraint

## Business Extension

Each factory has limited supply capacity.

A factory cannot ship more than its available capacity.

---

## General Constraint

For each factory f:

```Sum of x_fd over all depots d must be less than or equal to Supply_f```

In simple form:

```Total shipment from factory f to all depots <= factory capacity of f```

---

## Expanded Form

For Factory 1:

```x_Factory1_Depot1 + x_Factory1_Depot2 + x_Factory1_Depot3 <= 500```

For Factory 2:

```x_Factory2_Depot1 + x_Factory2_Depot2 + x_Factory2_Depot3 <= 450```

---

## Table Form

| Factory   | Constraint                                         |
| --------- | -------------------------------------------------- |
| Factory 1 | Total shipment from Factory 1 to all depots <= 500 |
| Factory 2 | Total shipment from Factory 2 to all depots <= 450 |

---

## Business Meaning

This constraint ensures that no factory ships more than its available production or supply capacity.

---

# Step 11: Depot Flow Balance Constraint

## Business Extension

A depot is an intermediate distribution point.

In this model, the depot does not create products and does not lose products.

So, for each depot:

```Total quantity entering depot = Total quantity leaving depot```

---

## General Constraint

For each depot d:

```Sum of x_fd from all factories f = Sum of y_dc to all customers c```

In simple form:

```Inflow into depot d = Outflow from depot d```

---

## Expanded Form

For Depot 1:

```x_Factory1_Depot1 + x_Factory2_Depot1 = y_Depot1_Customer1 + y_Depot1_Customer2 + y_Depot1_Customer3 + y_Depot1_Customer4```

For Depot 2:

```x_Factory1_Depot2 + x_Factory2_Depot2 = y_Depot2_Customer1 + y_Depot2_Customer2 + y_Depot2_Customer3 + y_Depot2_Customer4```

For Depot 3:

```x_Factory1_Depot3 + x_Factory2_Depot3 = y_Depot3_Customer1 + y_Depot3_Customer2 + y_Depot3_Customer3 + y_Depot3_Customer4```

---

## Table Form

| Depot   | Flow Balance Meaning                         |
| ------- | -------------------------------------------- |
| Depot 1 | Inflow from factories = Outflow to customers |
| Depot 2 | Inflow from factories = Outflow to customers |
| Depot 3 | Inflow from factories = Outflow to customers |

---

## Business Meaning

This constraint ensures proper conservation of flow.

In simple words:

```Whatever enters the depot must leave the depot```

---

# Step 12: Depot Capacity and Opening Constraint

## Business Extension

A depot can be used only if it is opened.

If a depot is closed, no product can flow through it.

If a depot is opened, product flow through it cannot exceed depot capacity.

This is the main linking constraint between the binary depot-opening decision and the continuous shipment quantity.

---

## General Constraint

For each depot d:

```Sum of y_dc over all customers c <= DepotCapacity_d × z_d```

In simple form:

```Total shipment from depot d to all customers <= Depot capacity of d × depot opening decision of d```

---

## Expanded Form

For Depot 1:

```y_Depot1_Customer1 + y_Depot1_Customer2 + y_Depot1_Customer3 + y_Depot1_Customer4 <= 400 × z_Depot1```

For Depot 2:

```y_Depot2_Customer1 + y_Depot2_Customer2 + y_Depot2_Customer3 + y_Depot2_Customer4 <= 350 × z_Depot2```

For Depot 3:

```y_Depot3_Customer1 + y_Depot3_Customer2 + y_Depot3_Customer3 + y_Depot3_Customer4 <= 300 × z_Depot3```

---

## How This Constraint Works

If Depot 1 is closed:

```z_Depot1 = 0```

then:

```Total flow from Depot 1 to all customers <= 400 × 0```

So:

```Total flow from Depot 1 to all customers <= 0```

This means no shipment can pass through Depot 1.

---

If Depot 1 is opened:

```z_Depot1 = 1```

then:

```Total flow from Depot 1 to all customers <= 400 × 1```

So:

```Total flow from Depot 1 to all customers <= 400```

This means Depot 1 can handle up to 400 units.

---

## Table Form

| Depot   | If Depot Is Closed | If Depot Is Opened |
| ------- | ------------------ | ------------------ |
| Depot 1 | Flow <= 0          | Flow <= 400        |
| Depot 2 | Flow <= 0          | Flow <= 350        |
| Depot 3 | Flow <= 0          | Flow <= 300        |

---

## Business Meaning

This is a fixed-charge network design constraint.

It ensures that the model pays the fixed opening cost before using a depot.

---

# Step 13: Customer Demand Constraint

## Business Extension

Every customer has a fixed demand.

The total quantity received by each customer from all depots must be equal to that customer’s demand.

---

## General Constraint

For each customer c:

```Sum of y_dc over all depots d = Demand_c```

In simple form:

```Total quantity received by customer c = demand of customer c```

---

## Expanded Form

For Customer 1:

```y_Depot1_Customer1 + y_Depot2_Customer1 + y_Depot3_Customer1 = 180```

For Customer 2:

```y_Depot1_Customer2 + y_Depot2_Customer2 + y_Depot3_Customer2 = 220 ```

For Customer 3:

```y_Depot1_Customer3 + y_Depot2_Customer3 + y_Depot3_Customer3 = 260```

For Customer 4:

```y_Depot1_Customer4 + y_Depot2_Customer4 + y_Depot3_Customer4 = 190```

---

## Table Form

| Customer   | Demand Constraint                    |
| ---------- | ------------------------------------ |
| Customer 1 | Total received from all depots = 180 |
| Customer 2 | Total received from all depots = 220 |
| Customer 3 | Total received from all depots = 260 |
| Customer 4 | Total received from all depots = 190 |

---

## Business Meaning

This constraint ensures full customer service.

Every customer receives exactly their required demand.

---

# Step 14: Non-Negativity and Binary Restrictions

## Business Extension

Shipment quantities cannot be negative.

Depot opening decisions must be yes/no decisions.

---

## Flow Non-Negativity

For all factories and depots:

```x_fd >= 0```

For all depots and customers:

```y_dc >= 0```

---

## Binary Depot Opening

For every depot:

```z_d must be either 0 or 1```

---

## Table Form

| Variable | Restriction                        |
| -------- | ---------------------------------- |
| x_fd     | Must be greater than or equal to 0 |
| y_dc     | Must be greater than or equal to 0 |
| z_d      | Must be either 0 or 1              |

---

## Business Meaning

The model cannot ship negative quantities.

Also, a depot cannot be half-opened. It must be either opened or closed.

---

# Step 15: Complete Mathematical Model in Readable Form

## Full Model

```Minimize Z```

where:

```Z = Sum over all factories and depots of factory-to-depot cost × factory-to-depot quantity + Sum over all depots and customers of depot-to-customer cost × depot-to-customer quantity + Sum over all depots of fixed depot opening cost × depot opening decision```

---

## Subject to the following constraints:

### 1. Factory Capacity

```For every factory: Total shipment from that factory to all depots must be less than or equal to that factory’s capacity```

Readable form:

```For each factory f: Sum of x_fd over all depots d <= Supply_f```

---

### 2. Depot Flow Balance

```For every depot: Total inflow from factories must be equal tototal outflow to customers ```

Readable form:

```For each depot d: Sum of x_fd over all factories f = Sum of y_dc over all customers c```

---

### 3. Depot Capacity and Opening

```For every depot: Total shipment from that depot to all customers must be less than or equal to depot capacity × depot opening decision```

Readable form:

```For each depot d: Sum of y_dc over all customers c <= DepotCapacity_d × z_d```

---

### 4. Customer Demand

```For every customer: Total quantity received from all depots must be equal to customer demand```

Readable form:

```For each customer c: Sum of y_dc over all depots d = Demand_c```

---

### 5. Non-Negativity

```All factory-to-depot shipment quantities must be greater than or equal to 0. All depot-to-customer shipment quantities must be greater than or equal to 0. ```

Readable form:

```x_fd >= 0, y_dc >= 0```

---

### 6. Binary Restriction

```Each depot opening decision must be either 0 or 1.```

Readable form:

```z_d is either 0 or 1```

---

# Step 16: Complete Model in Table Form

| Model Component        | Readable Mathematical Form                             | Business Meaning                                          |
| ---------------------- | ------------------------------------------------------ | --------------------------------------------------------- |
| Objective              | Minimize Z                                             | Minimize total supply chain cost                          |
| Factory-to-depot cost  | Sum of CostFD_fd × x_fd over all factories and depots  | Cost of shipping from factories to depots                 |
| Depot-to-customer cost | Sum of CostDC_dc × y_dc over all depots and customers  | Cost of shipping from depots to customers                 |
| Fixed depot cost       | Sum of FixedCost_d × z_d over all depots               | Cost of opening depots                                    |
| Factory capacity       | Sum of x_fd over all depots <= Supply_f                | Factory cannot ship beyond capacity                       |
| Depot flow balance     | Sum of x_fd entering depot = Sum of y_dc leaving depot | Depot inflow must equal depot outflow                     |
| Depot capacity/opening | Sum of y_dc from depot <= DepotCapacity_d × z_d        | Closed depot cannot ship; open depot has limited capacity |
| Customer demand        | Sum of y_dc received by customer = Demand_c            | Each customer demand must be satisfied                    |
| Non-negativity         | x_fd >= 0 and y_dc >= 0                                | Shipment quantity cannot be negative                      |
| Binary decision        | z_d is either 0 or 1                                   | Depot is either open or closed                            |

# Step 19: OR Scientist Interpretation

This model is a **fixed-charge two-echelon supply network design model**.

It is stronger than a simple transportation model because it includes both:

1. continuous shipment quantity decisions,
2. binary depot opening decisions.

That is why the model is classified as:

```Mixed Integer Linear Programming```

or:

```MILP```

# Step 20: Explanation


> “This project is a two-echelon supply network design model. The first echelon is factory-to-depot flow, and the second echelon is depot-to-customer flow. The model decides shipment quantities and depot opening decisions. The objective is to minimize total cost, including factory-to-depot transportation cost, depot-to-customer transportation cost, and fixed depot opening cost. The constraints ensure factory capacity, depot flow balance, depot capacity activation, and exact customer demand satisfaction. Since the model combines continuous flow variables with binary depot-opening variables, it is a mixed-integer linear programming model.”

---

# Step 21: One-Line Summary

```Minimum-Cost Supply Network = Optimal Flows + Optimal Depot Opening Decisions + Demand Satisfaction + Capacity Feasibility```

In simple words:

> This model designs the lowest-cost supply chain network by deciding which depots to open and how products should flow from factories to customers through those depots.

````python
import gurobipy as gp
from gurobipy import GRB
import pandas as pd
import matplotlib.pyplot as plt

# -----------------------------
# Data
# -----------------------------

factories = ["Factory_1", "Factory_2"]
depots = ["Depot_1", "Depot_2", "Depot_3"]
customers = ["Customer_1", "Customer_2", "Customer_3", "Customer_4"]

factory_capacity = {
    "Factory_1": 500,
    "Factory_2": 450
}

depot_capacity = {
    "Depot_1": 400,
    "Depot_2": 350,
    "Depot_3": 300
}

customer_demand = {
    "Customer_1": 180,
    "Customer_2": 220,
    "Customer_3": 260,
    "Customer_4": 190
}

fixed_depot_cost = {
    "Depot_1": 1200,
    "Depot_2": 1000,
    "Depot_3": 900
}

cost_factory_to_depot = {
    ("Factory_1", "Depot_1"): 4,
    ("Factory_1", "Depot_2"): 6,
    ("Factory_1", "Depot_3"): 9,
    ("Factory_2", "Depot_1"): 5,
    ("Factory_2", "Depot_2"): 4,
    ("Factory_2", "Depot_3"): 3
}

cost_depot_to_customer = {
    ("Depot_1", "Customer_1"): 3,
    ("Depot_1", "Customer_2"): 5,
    ("Depot_1", "Customer_3"): 7,
    ("Depot_1", "Customer_4"): 6,

    ("Depot_2", "Customer_1"): 4,
    ("Depot_2", "Customer_2"): 3,
    ("Depot_2", "Customer_3"): 5,
    ("Depot_2", "Customer_4"): 4,

    ("Depot_3", "Customer_1"): 7,
    ("Depot_3", "Customer_2"): 6,
    ("Depot_3", "Customer_3"): 3,
    ("Depot_3", "Customer_4"): 4
}

# -----------------------------
# Model
# -----------------------------

model = gp.Model("Supply_Network_Design")

# Flow from factories to depots
flow_fd = model.addVars(
    factories, depots,
    lb=0,
    vtype=GRB.CONTINUOUS,
    name="Factory_to_Depot_Flow"
)

# Flow from depots to customers
flow_dc = model.addVars(
    depots, customers,
    lb=0,
    vtype=GRB.CONTINUOUS,
    name="Depot_to_Customer_Flow"
)

# Depot opening binary variable
open_depot = model.addVars(
    depots,
    vtype=GRB.BINARY,
    name="Open_Depot"
)

# Objective function
model.setObjective(
    gp.quicksum(cost_factory_to_depot[f, d] * flow_fd[f, d]
                for f in factories for d in depots)
    +
    gp.quicksum(cost_depot_to_customer[d, c] * flow_dc[d, c]
                for d in depots for c in customers)
    +
    gp.quicksum(fixed_depot_cost[d] * open_depot[d]
                for d in depots),
    GRB.MINIMIZE
)

# Factory capacity constraints
for f in factories:
    model.addConstr(
        gp.quicksum(flow_fd[f, d] for d in depots) <= factory_capacity[f],
        name=f"Factory_Capacity_{f}"
    )

# Depot flow balance constraints
for d in depots:
    model.addConstr(
        gp.quicksum(flow_fd[f, d] for f in factories)
        ==
        gp.quicksum(flow_dc[d, c] for c in customers),
        name=f"Depot_Flow_Balance_{d}"
    )

# Depot capacity constraints linked with depot opening
for d in depots:
    model.addConstr(
        gp.quicksum(flow_dc[d, c] for c in customers)
        <= depot_capacity[d] * open_depot[d],
        name=f"Depot_Capacity_{d}"
    )

# Customer demand constraints
for c in customers:
    model.addConstr(
        gp.quicksum(flow_dc[d, c] for d in depots) == customer_demand[c],
        name=f"Customer_Demand_{c}"
    )

# Solve
model.optimize()

# -----------------------------
# Results
# -----------------------------

if model.status == GRB.OPTIMAL:
    print("Optimal solution found")
    print("Minimum total supply chain cost:", round(model.ObjVal, 2))

    depot_results = pd.DataFrame({
        "Depot": depots,
        "Opened": [open_depot[d].X for d in depots],
        "Outbound_Flow": [sum(flow_dc[d, c].X for c in customers) for d in depots],
        "Capacity": [depot_capacity[d] for d in depots]
    })

    fd_results = []
    for f in factories:
        for d in depots:
            if flow_fd[f, d].X > 0.0001:
                fd_results.append({
                    "From": f,
                    "To": d,
                    "Quantity": flow_fd[f, d].X
                })

    dc_results = []
    for d in depots:
        for c in customers:
            if flow_dc[d, c].X > 0.0001:
                dc_results.append({
                    "From": d,
                    "To": c,
                    "Quantity": flow_dc[d, c].X
                })

    fd_df = pd.DataFrame(fd_results)
    dc_df = pd.DataFrame(dc_results)

    display(depot_results)
    display(fd_df)
    display(dc_df)
else:
    print("No optimal solution found")
````

````python
if model.status == GRB.OPTIMAL:
    depot_results["Utilization_%"] = (
        depot_results["Outbound_Flow"] / depot_results["Capacity"] * 100
    )

    plt.figure(figsize=(8, 5))
    plt.bar(depot_results["Depot"], depot_results["Utilization_%"])
    plt.title("Depot Utilization Percentage")
    plt.xlabel("Depot")
    plt.ylabel("Utilization %")
    plt.ylim(0, 110)
    plt.axhline(y=100, linestyle="--", linewidth=1)
    plt.tight_layout()
    plt.show()
````

# Final Outcome:
“This project models a two-echelon supply chain network with factories, depots, and customers. The model decides which depots to open and how much product should flow through each route. It minimizes total logistics cost, including factory-to-depot shipping, depot-to-customer shipping, and fixed depot opening cost. This demonstrates network-flow modeling, facility activation decisions, and supply chain cost optimization.”
