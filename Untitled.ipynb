```python
import hexaly.optimizer as hx
import pandas as pd

# -----------------------------
# Data
# -----------------------------

factories = ["Factory_1", "Factory_2"]
depots = ["Depot_1", "Depot_2", "Depot_3"]
customers = ["Customer_1", "Customer_2", "Customer_3", "Customer_4"]

factory_capacity = {
    "Factory_1": 500,
    "Factory_2": 450
}

depot_capacity = {
    "Depot_1": 400,
    "Depot_2": 350,
    "Depot_3": 300
}

customer_demand = {
    "Customer_1": 180,
    "Customer_2": 220,
    "Customer_3": 260,
    "Customer_4": 190
}

fixed_depot_cost = {
    "Depot_1": 1200,
    "Depot_2": 1000,
    "Depot_3": 900
}

cost_factory_to_depot = {
    ("Factory_1", "Depot_1"): 4,
    ("Factory_1", "Depot_2"): 6,
    ("Factory_1", "Depot_3"): 9,
    ("Factory_2", "Depot_1"): 5,
    ("Factory_2", "Depot_2"): 4,
    ("Factory_2", "Depot_3"): 3
}

cost_depot_to_customer = {
    ("Depot_1", "Customer_1"): 3,
    ("Depot_1", "Customer_2"): 5,
    ("Depot_1", "Customer_3"): 7,
    ("Depot_1", "Customer_4"): 6,

    ("Depot_2", "Customer_1"): 4,
    ("Depot_2", "Customer_2"): 3,
    ("Depot_2", "Customer_3"): 5,
    ("Depot_2", "Customer_4"): 4,

    ("Depot_3", "Customer_1"): 7,
    ("Depot_3", "Customer_2"): 6,
    ("Depot_3", "Customer_3"): 3,
    ("Depot_3", "Customer_4"): 4
}

# -----------------------------
# Hexaly Model
# -----------------------------

with hx.HexalyOptimizer() as optimizer:
    model = optimizer.model

    # Decision variables
    flow_fd = {
        (f, d): model.float(0, 100000)
        for f in factories
        for d in depots
    }

    flow_dc = {
        (d, c): model.float(0, 100000)
        for d in depots
        for c in customers
    }

    open_depot = {
        d: model.bool()
        for d in depots
    }

    # -----------------------------
    # Objective Function
    # -----------------------------

    factory_to_depot_cost = model.sum(
        cost_factory_to_depot[f, d] * flow_fd[f, d]
        for f in factories
        for d in depots
    )

    depot_to_customer_cost = model.sum(
        cost_depot_to_customer[d, c] * flow_dc[d, c]
        for d in depots
        for c in customers
    )

    depot_opening_cost = model.sum(
        fixed_depot_cost[d] * open_depot[d]
        for d in depots
    )

    total_cost = factory_to_depot_cost + depot_to_customer_cost + depot_opening_cost

    model.minimize(total_cost)

    # -----------------------------
    # Factory Capacity Constraints
    # -----------------------------

    for f in factories:
        model.constraint(
            model.sum(flow_fd[f, d] for d in depots) <= factory_capacity[f]
        )

    # -----------------------------
    # Depot Flow Balance Constraints
    # -----------------------------

    for d in depots:
        model.constraint(
            model.sum(flow_fd[f, d] for f in factories)
            ==
            model.sum(flow_dc[d, c] for c in customers)
        )

    # -----------------------------
    # Depot Capacity and Opening Constraints
    # -----------------------------

    for d in depots:
        model.constraint(
            model.sum(flow_dc[d, c] for c in customers)
            <=
            depot_capacity[d] * open_depot[d]
        )

    # -----------------------------
    # Customer Demand Constraints
    # -----------------------------

    for c in customers:
        model.constraint(
            model.sum(flow_dc[d, c] for d in depots) == customer_demand[c]
        )

    # Close model before solving
    model.close()

    # Optional time limit
    optimizer.param.time_limit = 10

    # Solve
    optimizer.solve()

    # -----------------------------
    # Results
    # -----------------------------

    print("Hexaly solution completed")
    print("Minimum total supply chain cost:", total_cost.value)

    depot_results = pd.DataFrame({
        "Depot": depots,
        "Opened": [open_depot[d].value for d in depots],
        "Outbound_Flow": [
            sum(flow_dc[d, c].value for c in customers)
            for d in depots
        ],
        "Capacity": [depot_capacity[d] for d in depots]
    })

    fd_results = []
    for f in factories:
        for d in depots:
            if flow_fd[f, d].value > 0.0001:
                fd_results.append({
                    "From": f,
                    "To": d,
                    "Quantity": flow_fd[f, d].value
                })

    dc_results = []
    for d in depots:
        for c in customers:
            if flow_dc[d, c].value > 0.0001:
                dc_results.append({
                    "From": d,
                    "To": c,
                    "Quantity": flow_dc[d, c].value
                })

    fd_df = pd.DataFrame(fd_results)
    dc_df = pd.DataFrame(dc_results)

    display(depot_results)
    display(fd_df)
    display(dc_df)
```
